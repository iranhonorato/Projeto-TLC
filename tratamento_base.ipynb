```python
import pandas as pd
```

```python
df = pd.read_excel("../Base 5.1.xlsx")
df
```

```python
df_pdf = df[~df["PDF"].isin(["Site", "Bloqueado"])]
df_pdf
```

```python
df_block_no_resum = df[
    (df["PDF"] == "Bloqueado") &
    (df["Resumo"] == False)
]
df_block_no_resum
```

```python
df_block_w_resum = df[
    (df["PDF"] == "Bloqueado") &
    (df["Resumo"] == True)]
df_block_w_resum
```

```python
df_site = df[df["PDF"] == "Site"]
df_site
```

```python
print("Total", len(df))
print("pdf", len(df_pdf))
print("bloqueado sem resumo", len(df_block_no_resum))
print("bloqueado com resumo", len(df_block_w_resum))
print("site", len(df_site))
print("Verificando", len(df_pdf)+len(df_block_no_resum)+len(df_site)+len(df_block_w_resum))
```

```python
with pd.ExcelWriter("../Base 5.1.xlsx", mode="a", engine="openpyxl", if_sheet_exists="replace") as writer:
    df_pdf.to_excel(writer, sheet_name="PDFs_validos", index=False)
    df_block_no_resum.to_excel(writer, sheet_name="Bloqueado_sem_Resumo", index=False)
    df_block_w_resum.to_excel(writer, sheet_name="Bloqueado_com_Resumo", index=False)
    df_site.to_excel(writer, sheet_name="Sites", index=False)
```
