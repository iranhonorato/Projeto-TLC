```python
def inicializa_driver():
    service = Service()
    options = webdriver.ChromeOptions()
    return webdriver.Chrome(service=service, options=options)


driver = inicializa_driver()
pdf = []
resumo = []
resultados = []
i = 0
for url in df["Trabalho"]:

    try:
        # === 1.1 Se o link já é PDF direto ===
        if url.lower().endswith(".pdf"):
            pdf.append(url)
            resumo.append(False)
            continue
        

        # === 1.2 Verifica o tipo de arquivo / link
        resp = requests.head(url, allow_redirects=True)
        content_type = resp.headers.get("Content-Type", "").lower()
        
        if "pdf" in content_type:
            pdf.append(url)
            resumo.append(False)
            continue
    

        driver.get(url)
        time.sleep(2)  # espera pagina renderizar

        html = driver.page_source
        soup = BeautifulSoup(html, "html.parser")
        

        # === 2. Procurar bloco de Arquivos (PDFs) ===
        pdf_link = None
        pdf_status = "Não existe"

        for bloco in soup.find_all("div", class_="simple-view-element"):
            header = bloco.find("h2")
            if header and "Arquivos" in header.get_text():
                for a in bloco.find_all("a", href=True):
                    href = a["href"]
                    if "download" in href:  # só pega os links certos
                        pdf_link = urljoin(url, href)  # monta URL absoluta
                        pdf_status = "Disponível"
                        break
                    elif "request-a-copy" in href:
                        pdf_status = "Bloqueado"
                break

        # === 3. Procurar Resumo ===
        resumo_existe = False
        for bloco in soup.find_all("div", class_="simple-view-element"):
            header = bloco.find("h2")
            if header and "Resumo" in header.get_text():
                corpo = bloco.find("div", class_="simple-view-element-body")
                if corpo and corpo.get_text(strip=True):
                    resumo_existe = True
                break

       # === 4. Montar resultado ===
        if pdf_link:  # PDF liberado
            H = pdf_link
        elif pdf_status == "Bloqueado":  # PDF bloqueado → salva link da página
            H = pdf_status
        else:
            H = "Site"


        pdf.append(H)
        resumo.append(resumo_existe)
        print(i , " - " ,H, " --- ", resumo_existe)
        i += 1
    except Exception as e:
        
        pdf.append("Erro")
        resumo.append("Erro")
        print(i , " - ", "Erro  ---  Erro")
        i += 1


driver.quit()

df["PDF"] = pdf
df["Resumo"] = resumo

# Salvar como um novo arquivo (não sobrescreve o original)
df.to_excel("Base 5.0.xlsx", index=False)
```
